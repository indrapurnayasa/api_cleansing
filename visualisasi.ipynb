```python
import pandas as pd

df = pd.read_csv("data.csv", encoding="ISO-8859-1")
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

MENGHITUNG PERSENTASE HATE SPEECH DAN ABUSIVE

```python

category_count = df['HS'].sum(), df['Abusive'].sum()
plt.pie(category_count, autopct='%1.1f%%', labels = ["Hate Speech", "Abusive"])
plt.axis('equal')
plt.show()
```

TINGKAT HATE SPEECH WEAK, MODERATE, STRONG

```python


category_count = df['HS_Weak'].sum(), df['HS_Moderate'].sum(), df['HS_Strong'].sum()
label = ["HS_Weak", "HS_Moderate", "HS_Strong"]

plt.bar(label, category_count, color=['aqua', 'darkturquoise', 'darkcyan'])
plt.show()
```

JENIS HATE SPEECH RELIGION, RACE, PHYSICAL, GENDER, OTHER

```python
category_count = df['HS_Religion'].sum(), df['HS_Race'].sum(), df['HS_Physical'].sum(), df['HS_Gender'].sum(), df['HS_Other'].sum()
label = ['HS_Religion','HS_Race', 'HS_Physical', 'HS_Gender', 'HS_Other']
plt.bar(label, category_count)
plt.show()
```

PERSENTASE HATE SPEECH GROUP AND INDIVIDUAL

```python
category_hs = df['HS_Group'].sum(), df['HS_Individual'].sum()
plt.pie(category_hs, autopct='%1.1f%%', labels = ["HS_Group", "HS_Individual"])
plt.show()
```
